# Hello World Agent

_Creating a minimal (or ‘Hello world’) agent that responds to generic questions both in non-streaming and streaming mode._

```python
# Imports packages

from agent_framework.ollama import OllamaChatClient
from agent_framework.openai import OpenAIChatClient
```

```python
# Sets endpoints for Ollama models to be available over web requests.

OLLAMA_ENDPOINT="http://localhost:11434"
OLLAMA_ENDPOINT_OPENAI="http://localhost:11434/v1"
OLLAMA_MODEL="llama3.2:3b"
```

## Non-Streaming Mode
_The agents responds in non-streaming mode._

### Using Ollama-Native Chat Client

```python
# Initializes an Ollama chat client
ollama_chat_client = OllamaChatClient(
    model=OLLAMA_MODEL      # The language model that the client serves with
    )

# Creates an agent with Ollama as chat client.
agent = ollama_chat_client.as_agent(
    instructions="You are a helpful general assistant."
)
```

```python
result = await agent.run("What is the capital of France?")  # Agent executes the provided message.
print(result)
```

### Using OpenAI Compatible Chat Client

```python
# Initializes an OpenAI compatible chat client
openai_chat_client = OpenAIChatClient(
        model="llama3.2:3b",
        base_url="http://localhost:11434/v1",   # URL that the model is available at
        api_key="ollama",                       # Dummy value for OpenAI-compatible local endpoints
    )

# Creates an agent with a chat client compatible to OpenAI (but provided by Ollama)
agent = openai_chat_client.as_agent(
        instructions="You are a helpful general assistant.",
    )
```

```python
result = await agent.run("What is the capital of India?")
print(result)
```

## Streaming Mode
_The agents responds in streaming mode._

```python
async for chunk in agent.run(
    "Tell me about Agentic AI.", 
    stream=True     # Allows render responses as they get generated, reducing perceived wait times
    ):
    
    if chunk.text:
        print(chunk.text, end="", flush=True)

print()
```
