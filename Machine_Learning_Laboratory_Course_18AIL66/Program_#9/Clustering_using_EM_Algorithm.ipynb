# Clustering using Expectation-Maximization (EM) Algorithm

```python
# Imports required packages

import numpy as np
from sklearn.datasets import load_iris
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
```

```python
# Loads Iris data set
iris = load_iris()
```

## Exploratory Data Analysis (EDA)

```python
# Checks the keys associated with dataset object
print(iris.keys())
```

```python
# Checks the shape of the data
print(iris.data.shape)
```

```python
# Checks for the features and their location in the array
print(iris.feature_names)
```

```python
# Prints the target classes
print(iris.target_names)
```

```python
# Set petal length and petal width as features for modeling
X = iris.data[:, 2:]
```

```python
# Displays the actual clusters labelled with different colours

colormap = np.array(['blue', 'orange', 'green'])

plt.scatter(X[:,0], X[:,1], c=colormap[iris.target])
plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")
plt.title("Actual Clusters")
plt.show()
```

## Modeling

```python
# Initializes model with default hyperparameters
model = GaussianMixture(n_components = 3, random_state=42)
```

```python
# Fits the model (without label)
model.fit(X)
```

```python
# Performs predictions on training instances
predicted_labels = model.predict(X)
```

```python
# Sets colour map and markers to be used while displaying instances in a predicted cluster
colormap_prediction = np.array(['purple', 'cyan', 'red'])
markers = ['o', 'v', 's']

# Plots the instances in the predicted clusters
for i, l in enumerate(predicted_labels):
    plt.plot(X[i,0], X[i,1], color=colormap_prediction[l], marker=markers[l])    
plt.show()
```
